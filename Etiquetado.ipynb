```python
import os
```

```python
 # Ruta de la carpeta donde se encuentran las imágenes
folder_path = r"C:\Users\Sebas\Documents\Tesis\DataSetClasificado_Recortado\FASE 6"
```

```python
# Obtener la lista de archivos en la carpeta
file_list = os.listdir(folder_path)
```

```python
# Recorrer cada archivo y renombrarlo
for i, filename in enumerate(file_list):
    # Obtener la extensión del archivo
    _, ext = os.path.splitext(filename)
    
    # Construir el nuevo nombre de archivo
    new_filename = "FASE 6 - {}".format(i+1) + ext
    
    # Ruta completa del archivo antiguo y nuevo
    old_path = os.path.join(folder_path, filename)
    new_path = os.path.join(folder_path, new_filename)
    
    # Renombrar el archivo
    os.rename(old_path, new_path)
     # Imprimir el progreso
    print("Renombrando: {} -> {}".format(filename, new_filename))

print("Renombrado completado.")
```
